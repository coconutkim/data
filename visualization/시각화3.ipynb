```python
x = np.linspace(-4,4,100)
y1 = x**3
y2 = 10*x**2-1

plt.plot(x,y1,'o-r',x,y2,'*--b')
plt.xlim(-1,1) #  
plt.ylim(-3,3)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('my plot')
plt.legend(['$y1=x^3$','$y2=10x^2-1$'],loc = 'upper right')
plt.show()
```
