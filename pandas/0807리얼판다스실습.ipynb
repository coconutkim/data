```python
fuel = car.사용연료별
fuelrow = fuel[:4].tolist()
fuelrow
```

```python
licol = car.columns.tolist()
locatecol = licol[3:]
locatecol
```

```python
len(locatecol)
```

```python
car.서울.iloc[[i for i in range(0,48) if i%4==0]]
```

```python
gasol = car.iloc[[i for i in range(0,48) if i%4==0]][locatecol]
ss1 = gasol.sum()
diesel = car.iloc[[i for i in range(0,48) if i%4==1]][locatecol]
ss2 = diesel.sum()
lpg = car.iloc[[i for i in range(0,48) if i%4==2]][locatecol]
ss3 = lpg.sum()
etc = car.iloc[[i for i in range(0,48) if i%4==3]][locatecol]
ss4 = etc.sum()

predata = np.array([[ss1],[ss2],[ss3],[ss4]])
data = predata.reshape(4,18)
data
```

```python
data[0][1]
```

```python
final = pd.DataFrame(data=data,columns=[locatecol])
final
```

```python
final.index = fuelrow
final
```

```python
final2 = final.T
final2
```

```python
final2 = final2.drop(index='전국')
final2
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
final2.shape
```

```python
n = final2.shape[0]
n
```

```python
index = np.arange(n)
index
```

```python
w = 0.2
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.bar(index - w, final2['휘발유'], width=w, label='휘발유') # x축의 국어 그래프 위치가 -0.25 이동. 그래프 굵기 0.25
plt.bar(index, final2['경유'], width=w, label='경유') # 영어 그래프 생성. 그래프 굵기 0.25
plt.bar(index + w, final2['LPG'], width=w, label='LPG') # x축의 수학 그래프 위치가 + 0.25 이동. 그래프 굵기 0.25
plt.bar(index + 2*w, final2['기타연료'], width=w, label='기타연료')
plt.legend(ncol=4) # 범례 가로로 3개 출력
plt.xticks(index,final2.index,rotation=60)
plt.tight_layout()
plt.show()

# 오류: 튜플로 받아서 이름에 오류, 리스트로 수정해보기
```
