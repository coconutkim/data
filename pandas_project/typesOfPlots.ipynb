#   heat map

  (heat)   (map)  

              



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#  
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [2, 4, 6, 8, 10],
    'C': [5, 4, 3, 2, 1]
}

df = pd.DataFrame(data)

#  
correlation_matrix = df.corr()

#  
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#   
np.random.seed(42)
data = pd.DataFrame({
    'Column1': np.random.normal(0, 1, 1000),
    'Column2': np.random.normal(5, 2, 1000),
    'Column3': np.random.normal(-2, 1.5, 1000),
    'Column4': np.random.normal(3, 0.5, 1000)
})

#   KDE  ()
sns.kdeplot(data.values.flatten(), color='green', label='Overall KDE')

#  3   KDE  ()
sns.kdeplot(data['Column1'], color='blue', label='Column1')
sns.kdeplot(data['Column2'], color='blue', label='Column2')
sns.kdeplot(data['Column3'], color='blue', label='Column3')

#  
plt.legend()
plt.title('Kernel Density Estimation (KDE) Plot')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```
